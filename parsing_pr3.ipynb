```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
df = pd.DataFrame(columns=['Название','Адрес','Макс. оценка','Средняя оценка'])
sanatoriy = ['metallurg','zapolyarie','burgas','odissey','rus-avangard','znanie-adler','fregat','sputnik']
c=0
for i in sanatoriy:
    c+=1
    url = 'https://русские-курорты.рф/sochi/sanatory/'+i
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    rating = soup.find_all('a', class_='btn pull-right btn-primary')
    rating = np.array([float(i.text[-3:]) for i in rating])
    location = soup.find_all('a',class_='under', target='_blank')[0].text.replace('/xa0','').replace('Адрес:','')
    name = soup.find_all('span', itemprop="name")[3].text
    df.loc[c] = [name, location, rating.max(), rating.mean()]
```

```python
df = pd.DataFrame(columns = ['Name','Genre','Rating Critics','Rating Viewers'])
url = 'https://metarankings.ru/best-movies/'
res = requests.get(url)
soup = BeautifulSoup(res.content,'html.parser')
genre = [i.text.replace('Жанр:','') for i in soup.find_all('p')][2:]
rating1 = [i.text[1:4].replace('\nЗ','.0').replace('\nР','.0') for i in soup.find_all('div', class_="graph")][::2]
rating2 = [i.text[1:4].replace('\nЗ','.0').replace('\nР','.0') for i in soup.find_all('div', class_="graph")][1::2]
title = soup.find_all('a', class_="name")
i = 1
for t, g, r1, r2 in zip(title, genre, rating1, rating2):
    df.loc[i] = [t.get_text(), g, r1, r2]
    i += 1
```

```python
df = pd.DataFrame(columns = ['Name','Year','Rating'])
url = 'https://www.imdb.com/chart/top'
res = requests.get(url)
soup = BeautifulSoup(res.content,'html.parser')
year = soup.find_all('span', class_='secondaryInfo')
rating = soup.find_all('strong')
titles2 = [i.text for i in soup.find_all('td', class_="titleColumn")]
title = [i[16:-8].strip() for i in titles2]
i = 1
for t, y, r in zip(title, year, rating):
    df.loc[i] = [t, y.get_text(), r.get_text()]
    i += 1
```

```python
df = pd.DataFrame(columns = ['Name','Author','Rating'])
url = 'https://www.livelib.ru/genre/Триллеры/best'
res = requests.get(url)
soup = BeautifulSoup(res.content,'html.parser')
titles = soup.find_all('a', class_='brow-book-name with-cycle')
authors = soup.find_all('a', class_='brow-book-author')
rating = soup.find_all('span', class_='rating-value stars-color-orange')
i = 1
for t, a, r in zip(titles, authors, rating):
    df.loc[i] = [t.get_text(), a.get_text(), r.get_text()]
    i += 1
```
